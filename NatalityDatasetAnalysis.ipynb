```python
%load_ext google.cloud.bigquery
```

```python
%matplotlib inline
```

```python
%%bigquery high_states
SELECT
    state AS state,
    COUNT(is_male) AS birth_count
FROM `bigquery-public-data.samples.natality`
WHERE state IS NOT NULL
GROUP BY state
ORDER BY birth_count DESC
LIMIT 6
```

```python
high_states.plot(kind='bar', x='state', y='birth_count');
```

```python
%%bigquery mother_age
SELECT
    mother_age AS age,
    COUNT(mother_age) AS frequency
FROM `bigquery-public-data.samples.natality`
WHERE source_year BETWEEN 1998 AND 2008
GROUP BY age
ORDER BY age ASC
```

```python
mother_age.plot(x='age', y='frequency');
```

```python
mother_age.plot(kind='bar', x='age', y='frequency');
```

```python
%%bigquery avg_father_age
SELECT
    source_year AS year,
    AVG(father_age) AS avg_age
FROM `bigquery-public-data.samples.natality`
WHERE year BETWEEN 1998 AND 2008
GROUP BY year
ORDER BY year ASC
```

```python
avg_father_age.plot(x='year');
```

```python
avg_father_age.plot(kind='bar', x='year', y='avg_age');
```
